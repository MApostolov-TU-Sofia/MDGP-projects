```python
# Form absolute value in column
def absolute_value(x):
    return x if x >= 0.0 else -x

# Class to represent a graph 
class Graph:  
    def __init__(self):
        self.graph = dict()

    def generateGraph(self, df, path):        
        rowsN = len(df) # rows count
        colsN = len(df.columns) # cols count
        matrix = df.iloc[0:rowsN, 0:colsN]
        headerArr = matrix.columns

        for row in range(len(matrix)):
            whiteArr = [-1, -1, -1]
            largestNumArr = [0.0, 0.0, 0.0]
            for col in range(1, colsN):
                mVal = absolute_value(matrix.iloc[row, col])
                if (mVal == 1.0):
                    continue

                # Check if value is higher
                if (mVal > largestNumArr[2]):
                    if (mVal > largestNumArr[1]):
                        if (mVal > largestNumArr[0]):
                            largestNumArr[2] = largestNumArr[1]
                            largestNumArr[1] = largestNumArr[0]
                            largestNumArr[0] = mVal
                            whiteArr[2] = whiteArr[1]
                            whiteArr[1] = whiteArr[0]
                            whiteArr[0] = col
                        else:
                            largestNumArr[2] = largestNumArr[1]
                            largestNumArr[1] = mVal
                            whiteArr[2] = whiteArr[1]
                            whiteArr[1] = col
                    else:
                        largestNumArr[2] = mVal
                        whiteArr[2] = col

            # Companies array
            companiesArr = [headerArr[whiteArr[0]], headerArr[whiteArr[1]], headerArr[whiteArr[2]]]
            i = -1
            j = 0
            indexesToRemove = [0, 0, 0]
            # Remove duplicated companies and set largest numbers items to remove
            for x in companiesArr.copy():
                i = i + 1
                if (x in self.graph):
                    for y in self.graph[x]:
                        if (y == headerArr[row + 1]):
                            companiesArr.remove(x)
                            indexesToRemove[j] = largestNumArr[i]
                            j = j + 1

            # Remove largest numbers items
            for x in largestNumArr.copy():
                if (x in indexesToRemove):
                    largestNumArr.remove(x)

            values = dict()
            i = 0
            for x in companiesArr:
                values[x] = largestNumArr[i]
                i = i + 1
                
            # Add a new item in graph and edges dictionary
            self.graph[headerArr[row + 1]] = values
    
        # Return new formed graph
        return self.graph
```
