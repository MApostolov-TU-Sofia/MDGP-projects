```python
from operator import itemgetter

# Kruskal's algorithm to find Maximal Spanning Tree of a given connected, undirected and weighted graph  
class MaximalSpanningTreeGraph:  
    def __init__(self):
        self.graph = dict()
  
    # The main function to construct MST 
    # using Kruskal's algorithm 
    def KruskalMST(self, generatedGraph):
        edges = dict()
        i = 0
        for k in generatedGraph:
            for iK in generatedGraph[k]:
                edges[i] = {
                    'source': k,
                    'target': iK,
                    'weight': generatedGraph[k][iK]
                }
                i = i + 1
        
        # order by descending 'weight' value
        edgesDesc = dict(sorted(edges.items(), key=lambda t: t[1]['weight'], reverse=True))
#         print(edgesDesc)

        maximumSpanningTree = dict()
        visitedNodes = []
        for k in edgesDesc:
            if ((edgesDesc[k]['source'] not in visitedNodes) and (edgesDesc[k]['target'] in visitedNodes)):
                if (edgesDesc[k]['target'] not in maximumSpanningTree):
                    maximumSpanningTree[edgesDesc[k]['target']] = dict()                     
                    
                maximumSpanningTree[edgesDesc[k]['target']][edgesDesc[k]['source']] = edgesDesc[k]['weight']
                visitedNodes.append(edgesDesc[k]['source'])
                
            elif((edgesDesc[k]['source'] in visitedNodes) and (edgesDesc[k]['target'] not in visitedNodes)):
                if (edgesDesc[k]['source'] not in maximumSpanningTree):
                    maximumSpanningTree[edgesDesc[k]['source']] = dict()
                    
                maximumSpanningTree[edgesDesc[k]['source']][edgesDesc[k]['target']] = edgesDesc[k]['weight']
                visitedNodes.append(edgesDesc[k]['target'])
                
            elif ((edgesDesc[k]['source'] not in visitedNodes) and (edgesDesc[k]['target'] not in visitedNodes)):
                if (edgesDesc[k]['source'] not in maximumSpanningTree):
                    maximumSpanningTree[edgesDesc[k]['source']] = dict()
                    
                maximumSpanningTree[edgesDesc[k]['source']][edgesDesc[k]['target']] = edgesDesc[k]['weight']
                visitedNodes.append(edgesDesc[k]['source'])
                visitedNodes.append(edgesDesc[k]['target'])
            else:
#                 if (edgesDesc[k]['source'] in generatedGraph['graph'][edgesDesc[k]['source']].keys()):
                if (edgesDesc[k]['source'] not in maximumSpanningTree):
                    maximumSpanningTree[edgesDesc[k]['source']] = dict()
            
                maximumSpanningTree[edgesDesc[k]['source']][edgesDesc[k]['target']] = edgesDesc[k]['weight']
                    
        return maximumSpanningTree
  
```
